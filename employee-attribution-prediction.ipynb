# Employee Attrition Prediction using Machine Learning


## Dataset Loading
In this step, we load the employee dataset.


```python
import pandas as pd

df = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head()
```

## Data Understanding
We check the columns and structure of the dataset.



```python
df.columns
```

## Data Preprocessing
Here, we convert categorical values into numerical format for model training.


```python
df['Attrition'] = df['Attrition'].map({'Yes':1, 'No':0})
df = pd.get_dummies(df)
```

## Model Training
We use Logistic Regression to train the model and predict employee attrition.


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X = df.drop('Attrition', axis=1)
y = df['Attrition']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

## Data Visualization
Visualizing employee attrition and age distribution.


```python
import matplotlib.pyplot as plt

df['Attrition'].value_counts().plot(kind='bar')
plt.title("Attrition Count")
plt.show()
```

```python
plt.hist(df['Age'])
plt.title("Age Distribution")
plt.show()
```

## Conclusion
The model predicts employee attrition with good accuracy using machine learning techniques.
